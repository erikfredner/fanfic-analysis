```python
# class representing arbitrary text
# factory methods for class instantiating from other source
# constructors
class Text:
    
    @classmethod
    # helper functions producing instances of the class
    # cls = text as class of objects, not individual/self
    def from_file(cls, path):
        with open(path) as fh:
            # cls here calls the init function
            return cls(fh.read())
    
    def __init__(self, text):
        self.text = text   
```

```python
hp1 = Text.from_file('data/hp1.txt')
```
